```python
parent_dir = './outputs/SWIR/'
target_label = 'SWIR'
test_band = 90

reader_plotter = ImageReaderPlotter()
filenames_dict = reader_plotter.group_files_by_prefix(parent_dir)
```

```python
for target_apple in filenames_dict.keys():

    images = []
    gs_images = []

    for fn in filenames_dict[target_apple]:
        image = reader_plotter.read_image(parent_dir + fn, target_label, transpose=False)
        images.append(image)

    for image in images:
        gs_images.append(image[:, :, test_band])
    reader_plotter.plot_time_series_images(gs_images, filenames_dict[target_apple], test_band)
```

```python
# Read RGB images from png

parent_dir = '/home/sotgab/sysops/HIPPA/bruising/processed/RGB/'

reader_plotter = ImageReaderPlotter()
filenames_dict = reader_plotter.group_files_by_prefix(parent_dir, extension='png')

sampled_apples = random.sample(list(filenames_dict.keys()), 10)
sampled_apples
```

```python
for target_apple in sampled_apples:

    images = []

    for fn in filenames_dict[target_apple]:
        image = np.array(Image.open(parent_dir + fn))
        images.append(image)

    reader_plotter.plot_time_series_rgb_images(images, filenames_dict[target_apple])
```
